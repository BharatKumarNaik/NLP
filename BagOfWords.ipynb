```python
import pandas as pd
messages = pd.read_csv(r'Datasets/dataset.csv')
```

```python
messages.head()
```

```python
# Data Cleaning and Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages[:1000])):
    review = re.sub('[^a-zA-Z]',' ',messages['text'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[:10]
```

```python
# Create the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500, binary = False)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
# Create the Bag of Words model with N Grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500, binary = False,ngram_range=(1,2))
# ngram_range=(1,1) will be unigram alone, will be same as other BOW
# ngram_range=(1,2) will be combination unigram and bigram
# ngram_range=(1,3) will be combination of unigram, bigram and trigram
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
# It will include the combination of N neighboring words.
```
