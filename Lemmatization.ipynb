### Wordnet Lemmatizer
Lemmatization technique is like steeming. The output we will get after lemmatization is called "lemma", which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer calss which is a thin wrapper around the wordnet corpus. This class uses morpy() function to the WordNet CorpusReader class to find a lemma.

```python
import nltk
nltk.download()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python

lemmatizer = WordNetLemmatizer()
```

```python
'''
It takes POS as a argument
Noun - n
verb - v
adjective - a
adverb - r

by default POS tag is "n"
'''
print(lemmatizer.lemmatize("going"),lemmatizer.lemmatize("goes"))
```

```python
lemmatizer.lemmatize("going",pos="v"),lemmatizer.lemmatize("goes",pos="v")
```

```python
words = ['eating','eats','eaten','writing','writes','programming','fairly','sportingly']
for word in words:
    print(f"noun: {lemmatizer.lemmatize(word)}, verb: {lemmatizer.lemmatize(word,pos='v')}")
```
