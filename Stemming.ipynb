```python
words = ['eating','eats','eaten','writing','writes','programming','fairly','sportingly']
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
for word in words:
    print(f'{word}-->{stemming.stem(word)}')
```

### RegexpStemmer Class
NLTK has RegexStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regual expression and removes prefix or suffix that matches the expression.

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
# we need to provide regex for the stemming, and min is representing the min length of the word
# so in this case it will only stem for words with minimum length 4
# regex explained: where the word ending with ing or s or able remove those words.
```

```python
reg_stemmer.stem('eating')
```

### Snowball Stemmer
(Has better form of words then PorterStemmer)

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(f'{word} --->{stemmer.stem(word)}')
```

```python
# But its still not good enough
stemmer.stem("goes") #It's failing to convert it to proper root form
```
