```python
paragraph = '''
\nStop words are commonly occurring words in a language that usually carry little meaningful information on their own and are often removed during text preprocessing in NLP.\n Examples in English include words such as "the," "is," "and," "a," "of," and "in.\n" These words appear very frequently across documents but typically do not help distinguish the main topic or meaning of a text.\n Removing stop words can reduce the size of the vocabulary, decrease computational complexity, and improve the efficiency of NLP models by allowing them to focus on more informative words.\n For example, in the sentence "The cat is sitting on the mat," removing stop words leaves "cat sitting mat," which still captures the core meaning of the sentence.\n Stop word removal is especially useful in tasks such as information retrieval, text classification, document clustering, and search engines.\n However, stop words should not always be removed because they can be important in certain applications, such as sentiment analysis, machine translation, or question answering, where words like "not," "never," or "how" can significantly affect the meaning of a sentence.\n Therefore, the decision to remove stop words depends on the specific NLP task and its requirements.\n\n
'''
```

```python
print(paragraph)
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stopwords.words('english') 
# But if you notice stopwords will include some of the important words as well
# example: not, aren't isn't etc... these are indicating negativity in the sentence.
```

```python
stemmer = PorterStemmer()
```

```python
import nltk
sentences = nltk.sent_tokenize(paragraph)
```

```python
# Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) # word tokenization
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) # converting all the words into sentences.
    
```

```python
sentences
```

##### SnowballStemmer

```python
from nltk.stem import SnowballStemmer
sentences = nltk.sent_tokenize(paragraph)
stemmer = SnowballStemmer('english')
```

```python
# Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) # word tokenization
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) # converting all the words into sentences.
    
```

```python
sentences
```

#### Using Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
# Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) # word tokenization
    words = [lemmatizer.lemmatize(word,pos="v") for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) # converting all the words into sentences.
    
```
