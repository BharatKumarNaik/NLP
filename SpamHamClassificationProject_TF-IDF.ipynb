```python
import pandas as pd
messages = pd.read_csv(r'Datasets/dataset.csv')
messages.head()
```

```python
# Data Cleaning and Preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(messages[:500])):
    review = re.sub('[^a-zA-Z]',' ',messages['text'][i])
    review = review.lower()
    review = review.split()
    review = [wl.lemmatize(word,pos="n") for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
X = corpus
# Dependent Feature
y = pd.get_dummies(messages['label'][:500])
```

```python
y
```

```python
y = y.iloc[:,1].values #Taking either one column will do the job for considering o/p
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### Create TF-IDF

```python
# Create TF - IDF
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
# Independent Feature
X_train=cv.fit_transform(X_train).toarray()
```

```python
X_train
```

```python
cv.vocabulary_
```

```python
X_test = cv.transform(X_test).toarray()
X_test
```

### ML Model Training

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred)) #to check if the data is balanced or not
```
